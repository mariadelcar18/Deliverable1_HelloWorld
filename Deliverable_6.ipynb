```python
# Import the pandas, numpy, and seaborn libraries.
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Load the "iris" dataset and store it as a data frame.
iris = sns.load_dataset('iris')
```

```python
iris.head() # Show dataframe 
```

```python
# Drop the "species" column in the iris data frame.
iris.drop(labels = 'species', axis=1, inplace = True)
iris.head()
```

```python
#Add two more columns to the iris data frame
iris['sepal_sum'] = iris['sepal_length']+ iris['sepal_width']
iris['petal_sum'] = iris['petal_length']+ iris['petal_width']
iris.head()
```

```python
# Create a new data frame with the summary stats for the iris dataset
mean = list(iris.mean())
std = list(iris.std())
min = list(iris.min())
max = list(iris.max())

summary_dict = {'mean': mean, 'std':std, 'min':min, 'max':max}

summary_df = pd.DataFrame(data = summary_dict)
summary_df.rename(index={0: "sepal_length", 1: "sepal_width", 4: 'sepal_sum', 2: 'petal_length', 3: 'petal_width', 5: 'petal_sum'})
```
